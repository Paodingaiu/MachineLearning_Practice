```python
from cmath import log
import numpy as np
import pandas as pd
import logging
import matplotlib.pyplot as plt
```

```python
predata = pd.read_csv(r'../data/world-happiness-report-2017.csv',encoding='gb2312')
x1Name = "Economy..GDP.per.Capita."
x2Name = "Health..Life.Expectancy."
x3Name = "Freedom"
yName = "Happiness.Score"
```

## 查看原始数据趋势

```python
fig,ax = plt.subplots()
ax.scatter(predata.loc[:,x1Name],predata.loc[:,yName],label=x1Name,marker='o')
ax.scatter(predata.loc[:,x2Name],predata.loc[:,yName],label=x2Name,marker='x')
ax.scatter(predata.loc[:,x3Name],predata.loc[:,yName],label=x3Name,marker='*')
plt.ylabel(yName)
plt.legend()
plt.show()
```

## 初始化数据，执行特征缩放

```python
vectData = predata.loc[:,(x1Name,x2Name,x3Name,yName)]
dataMean = vectData.mean()
dataStd = vectData.std()
vectData = (vectData - dataMean)/dataStd
```

观察线性趋势

```python
fig,ax = plt.subplots()
ax.scatter(vectData.loc[:,x1Name],vectData.loc[:,yName],label=x1Name,marker='o')
ax.scatter(vectData.loc[:,x2Name],vectData.loc[:,yName],label=x2Name,marker='x')
ax.scatter(vectData.loc[:,x3Name],vectData.loc[:,yName],label=x3Name,marker='*')
plt.ylabel(yName)
```

观察数值在-2到2左右，可以接受梯度下降区间

准备数据 & 参数

```python
constantTheta = np.ones(vectData.shape[0])
data = vectData.loc[:,(x1Name,x2Name,x3Name)]
print(type(data))
data['constantTheta'] = constantTheta
yLabel = vectData.loc[:,yName]
print(type(yLabel))
theta = np.zeros((1,data.shape[1]))
learningRate = 0.1
iterations = 500
data
```

```python
def costFuncation(x,y,theta):
    return np.sum(np.power(x.dot(theta.T)-y,2))/(2*len(x))

costFuncation(data,yLabel,theta)
# data.dot(theta.T).add(yLabel,axis=1)
c = data.dot(theta.T)
c.sub(yLabel)
```
